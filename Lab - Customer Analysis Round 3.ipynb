## EDA (Exploratory Data Analysis) - Complete the following tasks to explore the data:

### Show DataFrame info.

```python
df.info()
```

### Describe DataFrame.

```python
df.describe().T.round(2) # I use the transpose method because I prefer how it looks and the round method to make numbers shorter
```

### Show a plot of the total number of responses.

```python
df["Response"].hist()
```

### Show a plot of the response rate by the sales channel.

```python
plt.figure(figsize=(10,6))
sns.countplot('Sales Channel', hue='Response', data=df)
plt.show()
```

### Show a plot of the response rate by the total claim amount.

```python
sns.boxplot(x="Response", y="Total Claim Amount", data=df)
plt.show()
```

### Show a plot of the response rate by income.

```python
plt.figure(figsize=(10,6))

sns.barplot(x="Response", y="Income", data=df)
plt.show()
```
